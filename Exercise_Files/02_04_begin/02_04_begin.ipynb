## Segment 4 - Concatenating and transforming data

```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

```python
np.random.seed(25)
DF_obj_1 = pd.DataFrame(np.random.rand(36).reshape(6,6))
DF_obj_1
```

```python
DF_obj_2 = pd.DataFrame(np.arange(15).reshape(5,3))
DF_obj_2
```

### Concatenating data

```python
pd.concat([DF_obj_1, DF_obj_2],axis=1)
```

```python
pd.concat([DF_obj_1, DF_obj_2])
```

### Transforming data
#### Dropping data

```python
DF_obj_1.drop([0,2])
```

```python
DF_obj_1
```

```python
DF_obj_1.drop([0,2], axis=1)
```

### Adding data

```python
series_obj = Series(np.arange(6))
series_obj.name = "added_variable"
series_obj
```

```python
variable_added = DataFrame.join(DF_obj_1, series_obj)
variable_added
```

```python
added_data_table = variable_added.append(variable_added, ignore_index=False)
added_data_table
```

```python
added_data_table = variable_added.append(variable_added, ignore_index=True)
added_data_table
```

### Sorting data

```python
DF_sorted = DF_obj_1.sort_values(by=(5), ascending=[False])
DF_sorted
```
